```python
import numpy as np
import plotly.graph_objects as go

def f(x, y):
    return x ** 2 + y ** 2

# разбиваем равномерно отрезки от -100 до 100 на 50 элементов
x = np.linspace(-100, 100, 50)
y = np.linspace(-100, 100, 50)

# создает двумерные массивы с сеткой для расчета значения z
x, y = np.meshgrid(x, y)
z = f(x, y)

# задаем фигуру
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-100, 100, 50)
y = np.linspace(-100, 100, 50)
x, y = np.meshgrid(x, y)
z = f(x, y)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание фигуры
fig = plt.figure()

# Создание 3D-осей
ax = fig.add_subplot(111, projection='3d')

# Создание сетки значений для x1 и x2
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)

# Вычисление значения функции в каждой точке сетки
Z = np.sin(X1 * X2)

# Создание цветового градиента
norm = plt.Normalize(Z.min(), Z.max())
colors = plt.cm.jet(norm(Z))

# Создание поверхности графика с цветовым градиентом
surf = ax.plot_surface(X1, X2, Z, facecolors=colors, shade=False)

# Настройка осей
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Z')

# Добавление цветовой шкалы
fig.colorbar(surf, shrink=0.5, aspect=5)

# Отображение графика
plt.show()
```
