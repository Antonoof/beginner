```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'London', 'Paris'],
        'Date': ['1980-10-13','1986-01-10','2005-02-23']}
df = pd.DataFrame(data)
df.to_csv('file.csv', index=True) # для сохранения в облаке
print(type(df))
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df = df.assign(Date_err=['1980,10-13','1986,01-10','2005,02-23']) # Создаем столбец
df
```

```python
df['Date_err'] = pd.to_datetime(df['Date_err'], format = '%Y,%m-%d') # Изменяем данные
df
```

```python
df = df.drop('Date_err', axis=1) # удаляем столбец
df
```

```python
df['City'] = ['New York', 'London', 'Russia'] # редактируем объект
df
```

```python
df.info()
df = df.astype({'Age' : float})
df.info()
```
