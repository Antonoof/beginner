# Типы нейронных сетей

В этой статье мы рассмотрим различные модели CNN (сверточных нейронных сетей), их архитектуру, а также их применение. Просмотрите список моделей CNN.

# Различные типы моделей CNN:

**1. LeNet:**

image.png

LeNet - самая популярная архитектура CNN, а также первая модель CNN, появившаяся в 1998 году. Изначально LeNet была разработана для классификации рукописных цифр от 0 до 9 в наборе данных MNIST. Он состоит из семи слоев, каждый из которых имеет свой собственный набор обучаемых параметров. Он принимает изображение размером 32 × 32 пикселя, что довольно велико по сравнению с изображениями в наборах данных, используемых для обучения сети. RELU - это функция активации, которая была использована. Слои выкладываются в следующем порядке:

• Первый сверточный слой состоит из 6 фильтров размером 5 X 5 с шагом 1.

• Второй слой представляет собой слой, объединяющий средние значения 2 X 2, или слой "подвыборки", с шагом 2.

• Третий слой также представляет собой сверточный слой с фильтрами размером 16,5x5 и шагом 1.

• Четвертый слой - это еще один слой среднего размера 2x2 с шагом 2.

• Пятый уровень в основном соединяет выходные данные четвертого уровня с полностью подключенным уровнем, состоящим из 120 узлов.

• Шестой уровень - это аналогичный полностью подключенный уровень с 84 узлами, который является производным от выходных данных 120 узлов Пятого уровня.

• Седьмой уровень состоит из классификации выходных данных предыдущего уровня по десяти классификациям на основе 10 цифр, которые он был обучен идентифицировать.

Это был один из первых эффективных алгоритмов распознавания рукописных цифр. Однако эта сеть оказалась неэффективной с точки зрения вычислительных затрат и точности, когда дело доходило до обработки огромных изображений и категоризации по большому количеству классов объектов.

# 2. AlexNet:

Начиная с ядра 11x11, Alexnet состоит из 5 уровней conv. Для трех массивных линейных уровней это был первый проект, в котором были использованы уровни максимального объединения, функции активации ReLU и отсева. Сеть была использована для классификации изображений по 1000 различным категориям.

Архитектура сети похожа на архитектуру Planet, но она включает в себя гораздо больше фильтров, чем в первоначальной версии Lender, что позволяет классифицировать гораздо больше объектов. Кроме того, она решает проблему переоснащения, используя "отсев", а не регуляризацию.

Базовая схема архитектуры AlexNet, показывающая ее пять сверточных и три полностью связанных уровня:

• Сначала применяется слой свертки (CL) с 96 фильтрами размером 11 X 11 и шагом 4.

• * После этого применяется слой максимального объединения (MPL) с размером фильтра 3 X 3 и шагом 2.

• Снова используется CL из 256 фильтров размером 5 X 5 и шагом = 4.

• * Затем был использован MPL с размером фильтра 3 X 3 и шагом 2.

• Была использована выборка из 384 фильтров размером 3 X 3 и шагом = 4.

• Опять же, выборка из 384 фильтров размером 3 X 3 и шагом = 4.

• Снова был использован CL из 256 фильтров размером 3 X 3 и шагом = 4.

• * Затем был использован MPl с размером фильтра 3 X 3 и шагом = 2.

• Когда выходные данные последнего слоя преобразуются во входной слой, как в Полностью связанном блоке, он содержит 9261 узел, каждый из которых полностью подключен к скрытому слою с 4096 узлами.

• Первый скрытый слой снова полностью связан со скрытым слоем, состоящим из 4096 узлов.

Для обучения исходной сети были использованы два графических процессора. Она содержит восемь уровней, каждый из которых имеет свой собственный набор настроек, которые можно изучить. В качестве входных данных для модели используются фотографии RGB. Relu - это функция активации, используемая на всех уровнях. Были использованы два выпадающих уровня. Softmax - это функция активации, используемая на уровне вывода.

# 3. ResNet:

ResNet - это хорошо известная модель глубокого обучения, которая впервые была представлена в статье Шаоцина Жэня, Каймина Хэ, Цзянь Суна и Сянъю Чжана. В 2015 году было опубликовано исследование под названием "Глубокое остаточное обучение для распознавания изображений". ResNet - одна из наиболее широко используемых и эффективных моделей глубокого обучения на сегодняшний день.

Для сброса используется так называемый остаточный блок.
Он основан на концепции "пропуска соединений" и использует множество методов пакетной нормализации, что позволяет успешно обрабатывать сотни слоев без ущерба для скорости с течением времени.

image.png

Первое, что мы замечаем на приведенной выше диаграмме, это наличие прямой связи, которая пропускает несколько уровней модели. "Пропускная связь", как известно, лежит в основе остаточных блоков. Из-за пропускной связи выходные данные не совпадают. Без подключения к пропуску входные данные 'X умножаются на веса слоя, а затем добавляется коэффициент смещения.

Архитектурный дизайн основан на VGG-19 и включает в себя 34-слойную простую сеть, к которой добавлены короткие и пропускные соединения. Как видно из приведенной ниже диаграммы, эти пропускные соединения или остаточные блоки преобразуют дизайн в остаточную сеть.

Идея заключалась в том, что на более глубоких уровнях обучения ошибок должно быть не больше, чем на более мелких. Чтобы воплотить эту идею в жизнь, были созданы переходные соединения. Создатели этой сети использовали вариант предварительной активации остаточного блока, в котором градиенты могут переходить по короткой ссылке на более ранние слои, сводя к минимуму проблему "исчезающих градиентов".

# 4. GoogleNet / Inception:

В конкурсе ILSVRC 2014 победила сеть GoogLeNet, или Inception Network, которая вошла в топ-5 по количеству ошибок, составив 6,67%, что практически соответствует уровню производительности человека. Google создала модель, которая включает в себя улучшенную реализацию оригинального дизайна LeNet. Это основано на концепции модуля inception. GoogLeNet - это разновидность сети Inception, которая представляет собой 22-слойную глубокую сверточную нейронную сеть.

В настоящее время GoogLeNet используется для различных приложений компьютерного зрения, включая распознавание лиц и идентификацию личности, обучение навыкам ведения боя и т.д.

Он состоит из 22 слоев (27 из которых объединены в пулы). После последнего начального модуля он использует глобальный средний пул.

# 5. MobileNetV1:

Девять модулей inception расположены друг над другом в архитектуре Inception Net/Googlenet, при этом между ними предусмотрено максимальное объединение слоев (для уменьшения пространственных размеров вдвое). Всего 22 слоя (27 из которых объединены в пулы). Он использует глобальный средний пул после окончательного начального модуля.

Модель мобильной сети построена на разделяемых по глубине свертках, которые представляют собой тип факторизованной свертки, которая делит обычную свертку на свертку по глубине и поточечную свертку. Свертка по глубине, используемая Mobilenetcz, применяет один фильтр к каждому входному каналу. Выходные данные свертки по глубине затем объединяются с помощью поточечной свертки с использованием свертки 11. За один шаг обычная свертка фильтрует и смешивает входные данные для создания нового набора выходных данных. Разделяемая по глубине свертка делит его на два уровня: один для фильтрации, а другой для объединения.
